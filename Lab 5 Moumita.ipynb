# Pick 5 companies to download the stock data from 01/01/2020 to 06/30/2020 and put them in a dataframe 

```python
stocks = ['DIS','FB','NFLX','SBUX','WMT']
```

```python
#download daily price data for each of the stocks in the portfolio
Daily_price = web.DataReader(stocks,data_source='yahoo',start='01/01/2020',end='06/30/2020')['Adj Close']
Daily_price
```

# a)	Plot a pricing chart for the duration of the timeline of all five stocks

```python
plt.figure(figsize=(15, 6))
for c in  Daily_price.columns.values:
    plt.plot( Daily_price.index,  Daily_price[c], lw=3, alpha=0.8,label=c)
plt.legend(loc='upper left', fontsize=14)
plt.ylabel('price in $')
```

# Explanation:
The grahs is plotted by using matplolob package.
Plotting is done after looping the vlaue of the column of the table. In the graph the x axis is the index or date and the y axis is the vlaue in the columns or price. Therefore tha graph is plotted against year and price of the stocks. The label for x axis is value in the date column. Lw is the line width in the graph. The legend are the defaulted best in the upper left corner with font size 14. The label for y axis is "price in $".
The over all figure size is 15,6, which describes the size of the graph.

# b)	Calculate daily returns of each stock and plot them in a chart.

```python
returns = Daily_price.pct_change()
plt.figure(figsize=(15, 6))
for c in returns.columns.values:
    plt.plot(returns.index, returns[c], lw=3, alpha=0.8,label=c)
plt.legend(loc='upper right', fontsize=14)
plt.ylabel('daily returns')
```

# Explanation:
In the above line of code. "pct_change()" is used to calculate the change i.e. between new vlaue form the old value.
The grahs is plotted by using matplolob package. Plotting is done after looping the value of the column of the table. In the graph the x axis is the index or date and the y axis is the vlaue in the columns or price of the return data. Therefore tha graph is plotted against year and daily returns of the stocks.
